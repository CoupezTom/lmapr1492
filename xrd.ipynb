### Dans l'énoncé, il nous est demandé de déterminer les trois premiers pics du diffractogramme qui correspondent à notre cristal en y indiquant les indices h k et l. On travaille avec un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Angström.

```python
#### Import ####

from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
#### On récupère la structure de notre matériaux. ####

####On crée notre stucture sur base du fichier MP. ####
    
    
with MPRester ("ozHTdwh3TX0utQzQ") as m:
   
    Structure = m.get_structure_by_material_id("mp-9364")
```

```python
#### On crée un XDR calculator pour notre diffractogramme. ####


xrd = XRDCalculator(1.54060)

#### On trouve le modèle de diffraction associé à notre cristal.####

Mod = xrd.get_pattern(Structure)

#### On imprime le diffractogramme.####

xrd.show_plot(Structure, annotate_peaks = False) 
#J'ai enlevé les indices correspondants aux pics car le graphe était illisible.
```

### On va pouvoir maintenant analyser nos trois premiers pics. Le diffractogramme nous montre l'angle $2\theta$ en fonction de l'intensité. 

# Premier pic

```python
print('2*theta : ' + str(Mod.x[0]) + ' degrés' )
print('Intensité : ' + str(Mod.y[0]))
print('Indices h k l : ' + str(Mod.hkls[0]))
```

# Deuxième pic

```python
print('2*theta : ' + str(Mod.x[1]) + ' degrés' )
print('Intensité : ' + str(Mod.y[1]))
print('Indices h k l : ' + str(Mod.hkls[1]))
```
